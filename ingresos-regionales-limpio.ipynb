```python
import pandas as pd
```

```python
data = pd.read_excel('datos_municipales_20211015002650_Sin-Corrección-Monetaria.xlsx')
data
```

```python
names = []
for a in data.keys():
    names.append(data[a][0])

dict = {}
cont = 0
for a in data.keys():
    if cont == 0 or cont == 1:
        dict[a] = ''
    else:
        dict[a] = names[cont]
    cont += 1
```

```python
data = data.rename(dict, axis = 'columns')
data.head()
```

```python
data = data.drop(0, axis = 'rows')
data
```

```python
# No hay datos nulos
cont = 0
for a in data.keys():
    for item in data[a].isna():
        if item == True:
            cont+=1
cont
    
```

```python
# No hay datos duplicados
cont = 0
for item in data.duplicated():
    if item == True:
        cont += 1
cont
```
